```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df_customer = pd.read_csv('Case Study - Customer.csv', delimiter=';')
df_product = pd.read_csv('Case Study - Product.csv', delimiter=';')
df_store = pd.read_csv('Case Study - Store.csv', delimiter=';')
df_transaction = pd.read_csv('Case Study - Transaction.csv', delimiter=';')
```

```python
merged_data = pd.merge(df_transaction, df_customer, on="CustomerID").merge(df_product, on="ProductID").merge(df_store, on="StoreID")
```

```python
merged_data
```

```python
merged_data.info()
```

```python
merged_data.describe()
```

```python
merged_data['Date'] = pd.to_datetime(merged_data['Date'], format='%d/%m/%Y')
```

```python
merged_data['Income'] = merged_data['Income'].str.replace(',', '.').astype(float)
```

```python
merged_data['Latitude'] = merged_data['Latitude'].str.replace(',', '.').astype(float)
```

```python
merged_data['Longitude'] = merged_data['Longitude'].str.replace(',', '.').astype(float)
```

```python
merged_data.to_csv("data_gabung.csv")
```
